```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = pd.read_csv(("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv"))
```

```python
x.head()
```

**Scatter Plot**

Uses dot to represent values for ***two different*** numeric variables.

Used to show relationship between variables.

```python
plt.scatter(x.crime_rate, x.percent_senior)
plt.show() #both must have same size
#Outliers can be observed
```

```python
##Adding titles and labels
plt.scatter(x.percent_senior, x.crime_rate)
plt.title('Plot of Crime Rate vs Percent Senior') 
plt.xlabel('Percent Senior')
plt.ylabel('Crime Rate')
```

**Line** **Chart**

Show trend over time.


```python
plt.plot(x.work_force,x.income) # X and Y
plt.xlabel('Work Force')
plt.ylabel('Income')
```

Change the size of the plot

```python
plt.figure(figsize=(12,5)) #12x5 plot
plt.plot(x.work_force,x.income) # X and Y
plt.xlabel('Work Force')
plt.ylabel('Income')
```

Formatting the style of your plot


```python
plt.plot(x.work_force,x.income, linestyle='--' , marker ='o', color ='r')
```

```python
#plt.plot(x.work_force,x.income, '--ro') #red o
#plt.plot(x.work_force,x.income, '--gx') #green x
#plt.plot(x.work_force,x.income, 'b^') #blue triangle
plt.plot(x.work_force, x.income, "go--", linewidth=3) # dashed lines of length 3
```

**Legend**
Area describing elements of a graph

```python
plt.plot(x.work_force,x.income, color = 'r', label = 'work_force')
plt.plot(x.physicians, x.income, label = 'physicians')
plt.legend
```

**Multiple Plots in one figure**

Using .subplot(nrows,ncols,index) method

'nrows' is the number of rows.
'ncols' is the number of columns in your figure.

```python
plt.figure(figsize=(12,5)) # sets size of all figures after the command
plt.subplot(1,2,1) # 1 row and 2 columns
plt.plot(x.work_force,x.income, 'go')
plt.title('Income vs Work Force')

plt.subplot(1,2,2)#.label_outer()
plt.plot(x.hospital_beds, x.income, 'r^')
plt.title('Income vs Hospital Beds')

plt.suptitle('Sub Plots') #Centred title to figure
```
